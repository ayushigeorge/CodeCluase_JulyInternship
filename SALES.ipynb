```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
encodings = ['utf-8', 'latin1', 'ISO-8859-1']

for encoding in encodings:
    try:
        data = pd.read_csv('retail.csv', encoding=encoding)
        break
    except UnicodeDecodeError:
        continue
```

```python
# Display the first few rows of the dataset
data.head()

# Display the summary statistics of the dataset
data.describe()

# Check the data types of each column
data.info()
```

```python
# Calculate total sales
total_sales = data['Quantity'] * data['UnitPrice']
total_sales = total_sales.sum()
```

```python
# Calculate average sales per transaction
average_sales = total_sales / data['InvoiceNo'].nunique()
```

```python
# Calculate total sales by product category
sales_by_category = data.groupby('Description')['Quantity'].sum()
```

```python
# Calculate total sales by month
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data['Month'] = data['InvoiceDate'].dt.month
sales_by_month = data.groupby('Month')['Quantity'].sum()
```

```python
# Plot the sales by category
sales_by_category.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Product Description')
plt.ylabel('Total Sales')
plt.title('Sales by Product Category')
```

```python
# Plot the sales by month
sales_by_month.plot(kind='line', figsize=(10, 6))
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Sales by Month')
```

```python
# Show the plots
plt.show()
```
